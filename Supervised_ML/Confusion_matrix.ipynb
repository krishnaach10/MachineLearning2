```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r"C:\Users\MEGHARAJ1\OneDrive\cmprogram\Desktop\MachineLearning_Revise\Supervised_ML\placement.csv")
```

```python
dataset.head()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x = "cgpa", y = "score", hue = "placement", palette = ["red", "blue"], data = dataset)
plt.xlabel("Cgpa")
plt.ylabel("Score")
plt.legend()
plt.show()
```

```python
x = dataset[["cgpa", "score"]]
y = dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.predict([[6.5,60]])
```

```python
lr.score(x_train, y_train), lr.score(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test, lr.predict(x_test))
cm
```

```python
sns.heatmap(cm, annot = True, fmt = "d")
plt.show()
```

```python
print(classification_report(y_test, lr.predict(x_test)))
```
