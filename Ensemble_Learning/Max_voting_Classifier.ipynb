```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=1000, noise= 0.2)
```

```python
df = pd.DataFrame(
    {
        "x1" : x[:,0],
        "x2" : x[0:,1],
        "y" : y
    }
)
```

```python
sns.scatterplot(x = "x1", y ="x2", hue = "y", data = df)
plt.show()
```

```python
x_a = df.iloc[:,:-1]
y_a = df["y"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_a, y_a, test_size = 0.2, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
dt.score(x_train, y_train) * 100, dt.score(x_test, y_test) * 100
```

```python
svc = SVC()
svc.fit(x_train, y_train)
svc.score(x_train, y_train) * 100 , svc.score(x_test, y_test) * 100 
```

```python
nb = GaussianNB()
nb.fit(x_train, y_train)
nb.score(x_train, y_train), nv.score(x_test, y_test)
```

```python
li = [("Dt", DecisionTreeClassifier()), ("SVC", SVC()), ("GNB", GaussianNB())]
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
vc = VotingClassifier(estimators = li)
vc.fit(x_train, y_train)
```

```python
vc.score(x_train, y_train), vc.score(x_test, y_test)
```

```python
pred = {
    "DT" : dt.predict(x_test),
    "SVC" : svc.predict(x_test),
    "GNB" : nb.predict(x_test),
    "VC" : vc.predict(x_test)
}
```

```python
com_pred = pd.DataFrame(pred)
```
