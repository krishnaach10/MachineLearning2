```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\MEGHARAJ1\OneDrive\cmprogram\Desktop\MachineLearning_Revise\Supervised_ML\cgpa_package_dataset.csv")
```

```python
df.head()
```

```python
x = df.drop("package", axis = "columns")
y = df["package"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
lr.score(x_train, y_train)*100, lr.score(x_test, y_test) * 100 
```

```python
svr = SVR(kernel = "linear")
svr.fit(x_train, y_train)
svr.score(x_train, y_train) *100, svr.score(x_test, y_test) * 100
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
dt.score(x_train, y_train)*100, dt.score(x_test, y_test)*100
```

```python
li = [("dt", DecisionTreeRegressor()), ("SVR", SVR(kernel = "linear")), ("LR", LinearRegression())]
```

```python
from sklearn.ensemble import VotingRegressor
vc = VotingRegressor(estimators = li)
vc.fit(x_train, y_train)
vc.score(x_train, y_train), vc.score(x_test, y_test)
```

```python
pred = {
    "DT" : dt.predict(x_test),
    "SVR" : svr.predict(x_test),
    "LR" : lr.predict(x_test),
    "VC" : vc.predict(x_test)
}
```

```python
predict_df = pd.DataFrame(pred)
```
